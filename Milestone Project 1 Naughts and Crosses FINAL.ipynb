```python
def start_game():
    
    print("Welcome to Naughts and Crosses. Not tic tac toe ;)")
    print("You can use numbers to say where you want to place your X or O in the grid.")
    print("The numbers correspond to the grid positions like this:\n")
    
    
    print("1 | 2 | 3")
    print("─────────")
    print("4 | 5 | 6")
    print("─────────")
    print("7 | 8 | 9")
    print("─────────")
    
```

```python
def select_player_marker():
    
    global player_1
    global player_2
    global acceptable_choices_marker
    
    player_1 = 'undecided'
    player_2 = 'undecided'
    acceptable_choices_marker = ["x","o"]
    
    
    while player_1 == "undecided":
        player_marker_choice = input("Do you want to be X or O?")
        
        if player_marker_choice.lower() in acceptable_choices_marker:
            player_1 = "player_" + str(player_marker_choice.lower())
                     
        else:
            print("Please only choose X or O")
            continue
            
        if player_1 == "player_x":
            player_2 = "player_o"
        
        else:
            player_2 = "player_x"
        
        
    print(f"OK, you'll be {player_1} and your opponent will be {player_2}.")
```

```python
from random import randint


def who_goes_first():
    
    global current_player
    
    current_player = 'undecided'
    
    if randint(1,2) %2 == 0:
        current_player = 'player_o'
        print("O goes first!")
    
    else:
        current_player = 'player_x'
        print("X goes first!")       
        
```

```python
def player_move():
    
    
    global player_position_choice
    global play_position
    global space_free
    global valid_player_position_choice
    global within_board_list_range
    global acceptable_board_list_range
    
    player_position_choice = "undecided"
    play_position = "undecided"
    space_free = False
    valid_player_position_choice = False
    within_board_list_range = False
    acceptable_board_list_range = range(0,10)
    
    #isdigit will check if a string contains numbers. User input is always strings.     
        
    while player_position_choice.isdigit() == False or within_board_list_range == False or space_free == False:
        player_position_choice = input("Please choose a grid position to take ")
                  
        #checks if the player provides a number    
        if player_position_choice.isdigit() == False:
            print("Please only choose an integer number ")
            continue
           
        
        #checks if the number is int he range
        if int(player_position_choice) in acceptable_board_list_range:
                within_board_list_range = True
                
        else:
            print("Please only choose a number between 1 and 9 ") 
            continue
           
        
        #checks if the space the player wants to use is free
        if board_list[int(player_position_choice)] == "":
            space_free = True
            
        else:
            print("Sorry that space is taken, please choose another")
            continue
        
        
        #sets the played position as the players choice, provided all criteria met
        if player_position_choice.isdigit() == True and within_board_list_range == True and space_free == True:
            play_position = int(player_position_choice)   
            
            
    return play_position
```

```python
def update_board():
    
    global current_player
    global play_position
    
    if current_player == "player_x":
        board_list[play_position] = "x"       
        
        
    if current_player == "player_o":
        board_list[play_position] = "o" 
        
```

```python
def game_won():
    
    global game_over  
    global board_list
    
    win_options_x = False
    win_options_o = False

    
    if ((board_list[1] == board_list[2] == board_list[3] == 'x') or 
    (board_list[4] == board_list[5] == board_list[6] == 'x') or 
    (board_list[7] == board_list[8] == board_list[9] == 'x') or
    (board_list[1] == board_list[4] == board_list[7] == 'x') or
    (board_list[2] == board_list[5] == board_list[8] == 'x') or
    (board_list[3] == board_list[6] == board_list[9] == 'x') or
    (board_list[1] == board_list[5] == board_list[9] == 'x') or
    (board_list[3] == board_list[5] == board_list[7] == 'x')):
        win_options_x = True
    
    elif ((board_list[1] == board_list[2] == board_list[3] == 'o') or
    (board_list[4] == board_list[5] == board_list[6] == 'o') or
    (board_list[7] == board_list[8] == board_list[9] == 'o') or
    (board_list[1] == board_list[4] == board_list[7] == 'o') or
    (board_list[2] == board_list[5] == board_list[8] == 'o') or
    (board_list[3] == board_list[6] == board_list[9] == 'o') or
    (board_list[1] == board_list[5] == board_list[9] == 'o') or
    (board_list[3] == board_list[5] == board_list[7] == 'o')):
        win_options_o = True
            

    
    
    if win_options_x == True:
        game_over = True
        print("Congratulations player_x wins!")
        
    elif win_options_o == True:
        game_over = True
        print("Congratulations player_o wins!") 
        
## needs a check if the board is full so no winners

# if count of "" (blank) in board list = 0, no winners but the board is full so the game needs to end
    elif board_list.count("") == 0:
        game_over = True
        print("No winner this time")
        
    else:
        print("Next move")
            
```

```python
def pass_play():
    
    global current_player
    
    
    #clear_output()
        
    if current_player == "player_x":
        current_player = "player_o"
        print("Please pass the game to player_o")
        
    else:
        current_player = "player_x"
        print("Please pass the game to player_x") 
    
```

```python
def play_on():
    
    global play_again
    global acceptable_choices_play_again
    
    play_again = "undecided"
    acceptable_choices_play_again = ["yes","no"]
    
    while play_again == "undecided":
    
        play_again_choice = input("Do you want to play again?")
    
        if play_again_choice.lower() not in acceptable_choices_play_again:
            print("Sorry I didn't understand, please type only 'yes' or 'no'")
            
        else:
            play_again = play_again_choice.lower()
            
    if play_again == "yes":
        print("OK, let's play again")
        
    else:
        print("Thanks for playing")    
```
